```python
from pdferret import PDFerret
import glob
import json
from dataclasses import asdict
```

```python
# instantiate it like this:
# for now only text_extractor supports two options, either "grobid" or "unstructured"
extractor = PDFerret(text_extractor="unstructured")
```

```python
batch = glob.glob("./test_data/Bond-IQ papers archive/**/*.pdf")
```

```python
%%time
res = extractor.extract_batch(batch)
```

```python
for doc in res:
    print(json.dumps(asdict(doc.metainfo), indent=4))
    print("-----------------")
```

```python
for chunk in doc.chunks:
    print(json.dumps(asdict(chunk), indent=4))
    print("-----")
```
